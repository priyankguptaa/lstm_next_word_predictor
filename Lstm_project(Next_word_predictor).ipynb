```python
data = """The sun sets behind the mountains.
She picked a flower from the garden.
He forgot his keys on the table.
They walked slowly through the forest.
I love reading books at night.
The dog chased the ball across the yard.
She smiled when she saw the surprise.
He opened the window to let in air.
We watched the stars from the rooftop.
The baby laughed at the funny face.
He poured milk into the glass.
She painted the wall a bright blue.
They built a sandcastle on the beach.
I heard a knock at the door.
The teacher wrote on the chalkboard.
He ran quickly to catch the bus.
She wore a red dress to the party.
The cat slept on the warm blanket.
He played guitar by the campfire.
They danced under the moonlight.
I found a coin on the sidewalk.
She baked cookies for the school fair.
He fixed the broken chair with glue.
We listened to music all night.
The bird flew out of the cage.
She drew a picture of a tree.
He climbed the ladder carefully.
They sang songs around the fire.
I spilled juice on the carpet.
She opened the book to chapter five.
He turned off the lights and left.
We planted flowers in the backyard.
The wind blew through the open window.
She tied her shoes before running.
He watched the rain from the porch.
They shared a pizza for lunch.
I packed my bag for the trip.
She whispered a secret in my ear.
He cleaned the dishes after dinner.
We played cards on the kitchen table.
The baby crawled across the floor.
She lit a candle on the cake.
He sharpened his pencil before class.
They waited patiently in the line.
I wrote a letter to my friend.
She brushed her hair in the mirror.
He counted the stars in the sky.
We rode our bikes to the park.
The phone rang during the meeting.
She folded the clothes neatly.
He smiled and waved goodbye."""
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts([data])
```

```python
len(tokenizer.word_index)
```

```python
input_sequence = []
for sentence in data.split('.'):
  tokenized_sentence = tokenizer.texts_to_sequences([sentence])[0]

  for i in range(1,len(tokenized_sentence)):
    input_sequence.append(tokenized_sentence[:i+1])
```

```python
input_sequence
```

```python
max_len = max([len(x) for x in input_sequence])
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequence = pad_sequences(input_sequence, maxlen=max_len,padding='pre')
```

```python
padded_input_sequence
```

```python
X = padded_input_sequence[:,:-1]
```

```python
y = padded_input_sequence[:,-1]
```

```python
X.shape
```

```python
y.shape
```

```python
from tensorflow.keras.utils import to_categorical
y = to_categorical(y,num_classes = 190)
```

```python
y.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,LSTM,Dense
```

```python
model = Sequential()
model.add(Embedding(190,100,input_length=7))
model.add(LSTM(90))
model.add(Dense(190,activation='softmax'))

model.build(input_shape=(None, 7))
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
y
```

```python
model.fit(X,y,epochs=250)
```

```python
import numpy as np
text = "baby"
for i in range(5):
  token_text = tokenizer.texts_to_sequences([text])[0]
  padded_token_text = pad_sequences([token_text],maxlen = 7,padding="pre")
  pos = np.argmax(model.predict(padded_token_text))

  for word,index in tokenizer.word_index.items():
    if index == pos:
      text = text + " " + word
      print(text)
```
